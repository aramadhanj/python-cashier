```python
from supermarket import Cashier
```

```python
def main():
    cashier = Cashier()
    while True:
        print("1. Add item")
        print("2. Update item quantity")
        print("3. Update item name")
        print("4. Update item price")
        print("5. Delete item")
        print("6. Check order")
        print("7. Calculate total price")
        print("8. Reset transaction")
        print("9. Exit")
        choice = input("Enter choice: ")
        if choice == "1":
            name = input("Enter item name: ")
            qty = int(input("Enter item quantity: "))
            price = float(input("Enter item price: "))
            cashier.add_item(name, qty, price)
            print("Item added successfully.")
        elif choice == "2":
            name = input("Enter item name: ")
            qty = int(input("Enter new quantity: "))
            if cashier.update_item_qty(name, qty):
                print("Item quantity updated successfully.")
            else:
                print("Item not found.")
        elif choice == "3":
            old_name = input("Enter current item name: ")
            new_name = input("Enter new item name: ")
            if cashier.update_item_name(old_name, new_name):
                print("Item name updated successfully.")
            else:
                print("Item not found.")
        elif choice == "4":
            name = input("Enter item name: ")
            price = float(input("Enter new price: "))
            if cashier.update_item_price(name, price):
                print("Item price updated successfully.")
            else:
                print("Item not found.")
        elif choice == "5":
            name = input("Enter item name: ")
            if cashier.delete_item(name):
                print("Item deleted successfully.")
            else:
                print("Item not found.")
        elif choice == "6":
            order = cashier.check_order()
            print(order)
        elif choice == "7":
            total_harga = sum(qty*price for qty, price in cashier.order_list.values())
            total_harga_setelah_diskon = cashier.total_price(total_harga)
            print(f"Harga sebelum diskon: {total_harga:,.0f}")
            print(f"Harga setelah diskon: {total_harga_setelah_diskon:,.0f}")
        elif choice == "8":
            cashier.reset_transaction()
            print("Transaction reset successfully.")
        elif choice == "9":
            print("Thank you for using the self-service cashier.")
            break
        else:
            print("Invalid choice. Please try again.")
```

```python
if __name__ == "__main__":
    main()
```
